## Will your car be stolen?

We have the following empirical data on car theft.

|Number | Color | Type | Origin | Stolen |
--------|-------|------|--------|--------|
| 1 | red | sport | domestic | yes |
| 2 | red | sport | domestic | no  |
| 3 | red | sport | domestic | yes |
| 4 | yellow | sport | domestic | no |
| 5 | yellow | sport | import | yes |
| 6 | yellow | station wagon | import | no |
| 7 | yellow | station wagon | import | yes |
| 8 | yellow | station wagon | domestic | no |
| 9 | red | combi | import | no |
| 10 | red | sport | import | yes |

Use the naïve Bayesian classifier to determine if there is a risk that the *red station wagon
domestic production* will be stolen.

&prop; is read "proportional to"

* P(yes $\mid$ color=red, type = station_wagon, origin=domestic) &prop; P(red | yes) P(station wagon $\mid$ yes)P(domestic | yes)P(yes) 
* P(no | color=red, type = station_wagon, origin=domestic) &prop; P(red|no) P(station wagon | no)P(domestic | no)P(no) 

ROC  dependency between TP(vertical) vs FP(horizintal axis)

```python
P_yes = 1/2
P_red_given_yes = (3/10) /P_yes             # P_red_given_yes = P_red_and_yes / P_yes
P_wagon_given_yes = (1/10)/P_yes             # P_wagon_given_yes = P_wagon_and_yes / P_yes
P_domestic_given_yes = (2/10)/P_yes        # P_domestic_given_yes = P_domestic_and_yes / P_yes

P_yes_given_red_wagon_domestic = P_red_given_yes*P_wagon_given_yes*P_domestic_given_yes*P_yes

P_no = 1-P_yes
P_red_given_no = (2/10)/P_no              # P_red_given_no = P_red_and_no / P_no
P_wagon_given_no = (2/10)/P_no             # P_wagon_given_no = P_wagon_and_no / P_no
P_domestic_given_no = (3/10)/P_no          # P_domestic_given_no = P_domestic_and_no / P_no

P_no_given_red_wagon_domestic = P_red_given_no*P_wagon_given_no*P_domestic_given_no*P_no

print("P_yes_given_red_wagon_domestic", P_yes_given_red_wagon_domestic)
print("P_no_given_red_wagon_domestic", P_no_given_red_wagon_domestic)

# EXPECTED
# P_yes_given_red_wagon_domestic 0.024
# P_no_given_red_wagon_domestic 0.04800000000000001
```

Classify the patient which

How does the calculation of the naïve Bayes classifier change if we use Laplace correction. Without this correction
$$P (E_k \mid H_t) = P(\text{sample from class } t \text{ satisfying } E_k) = \frac{\text{Number of patterns from } t \text{ satisfying } E_k}{\text{the number of pattern from } t}$$

with Laplace correction
$$P (E_k \mid H_t) = P(\text{sample from class } t \text{ satisfying } E_k) = \frac{\text{Number of patterns from } t \text{ satisfying } E_k + mp}{(\text{the number of patterns from } t) + m}$$

where $m$ is the number of artificially added patterns and $p$ is the a priori probability
$P(E_k \mid H_t)$. Very often $m = T$ and $p = \frac{1}{T}$ are used, where $T$ is the number of classes (different values) for the $k$-th attribute.

We will use the dataset for heart disease.

```python
%%file heart1.csv
Vitamin,    Size,   Excercise,   Bypass, Survived
B complex,  large,  regularly,  no,     -
B complex,  large,  regularly,  yes,    - 
C,          large,  regularly,  no,     + 
D,          medium, regularly,  no,     + 
D,          small,  little,     no,     + 
D,          small,  little,     yes,    - 
C,          small,  little,     yes,    +
B complex,  medium, regularly,  no,     - 
B complex,  small,  little,     no,     + 
D,          medium, little,     no,     +
B complex,  medium, little,     yes,    +
C,          medium, regularly,  yes,    + 
C,          medium, little,     no,     +
D,          big,    regularly,  yes,    -
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn.naive_bayes import CategoricalNB
```

Let us read the dataset.

```python
df = pd.read_csv("heart1.csv", sep="\s*,\s*", engine='python')
df
```

```python
X_raw = df.drop(columns=['Survived'])
y_raw = df['Survived'].values
y_raw
```

```python
from sklearn.preprocessing import OrdinalEncoder

enc_X = OrdinalEncoder()
X = enc_X.fit_transform(X_raw)
enc_y = OrdinalEncoder(categories=[['-','+']])
y = enc_y.fit_transform(y_raw.reshape(-1,1))
X
```

```python
enc_X.categories_
```

```python
y
```

```python
enc_y.categories_
```

```python
nbc = CategoricalNB()
nbc.fit(X, y)
```

```python
#Takes vitamin D, has a large family, exercises regularly and had a baypass
raw_res = nbc.predict(np.array([[2,1,1,1]]))
raw_res
```

```python
#Takes vitamin D, has a big family, exercises regularly and had a baypass
raw_res = nbc.predict([[2,0,1,1]])
raw_res
```

The returned value is the value of the encoded target attribute. We must use inverse transformation to obtain the result in the original encoding.

```python
res = enc_y.inverse_transform(raw_res.reshape(-1,1))
res
```

Predict survival status for a patient that takes vitamin C, has a large family and excercises regularly.

```python
enc_X1 = OrdinalEncoder()
X1 = enc_X1.fit_transform(X_raw.drop(columns=['Bypass']))
y = enc_y.fit_transform(y_raw.reshape(-1,1))
X1
```

```python
nbc1 = CategoricalNB()
nbc1.fit(X1, y)
```

```python
res1 = nbc1.predict(enc_X1.transform([['C', 'big', 'regularly']]))  #Takes vitamin D, has a small family, exercises regularly.
enc_y.inverse_transform(res1.reshape(-1,1))
```

Build and plot the corresponding decision tree. Will the decision tree classify the patient that takes vitamin C, has a large family and excercises regularly the same way as the Naïve Bayes classifier?

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

clf = DecisionTreeClassifier(random_state=0, criterion='entropy')
clf.fit(X1, y)

plt.figure(figsize=(8,8))
tree.plot_tree(clf, filled=True, fontsize=12)
```

```python
X_test = np.array([x for x in itertools.product(*enc_X.categories_)])
# print(X_test)
# print(enc_X.transform(X_test))
# print(nbc.predict(enc_X.transform(X_test)))
# print(nbc.predict(enc_X.transform(X_test)).reshape(-1,1))
NB_classification = enc_y.inverse_transform(nbc.predict(enc_X.transform(X_test)).reshape(-1,1))
NB_class_prob = nbc.predict_proba(enc_X.transform(X_test))
#print(NB_classification)
# print(NB_class_prob)
DT_classification = enc_y.inverse_transform(dtc.predict(enc_X.transform(X_test)).reshape(-1,1))
DT_class_prob = dtc.predict_proba(enc_X.transform(X_test))
#print(DT_classification)
results = pd.DataFrame(np.hstack((X_test, NB_classification, NB_class_prob, DT_classification, DT_class_prob)), 
                       columns=list(X_raw.columns)+['NB_class', 'NB_class_prob0', 'NB_class_prob1', 'DT_class', 'DT_class_prob0', 'DT_class_prob1'])
results
```
